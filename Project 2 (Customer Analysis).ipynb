### Churn Percentage

```python
churn_rate = df['Exited'].mean() * 100
print("Churn Rate:", churn_rate)
```

## GROUPBY ANALYSIS

#### Churn by Geography

```python
geo_churn = df.groupby('Geography')['Exited'].mean() * 100
print(geo_churn)
```

```python
geo_churn.plot(kind='bar',color=["darkblue", "orange", "darkred"])
plt.title('Churn Rate by Geography',fontsize=14,fontweight='bold')
plt.ylabel('Churn Percentage',fontsize=12,fontweight='bold')
plt.xlabel('Geography',fontsize=12,fontweight='bold')
plt.show()
```

#### Churn by Gender

```python
gender_churn = df.groupby('Gender')['Exited'].mean() * 100
gender_churn
```

```python
gender_churn.plot(kind='bar',color=["pink","k"])
plt.title('Churn Rate by Gender',fontsize=14,fontweight='bold')
plt.ylabel('Churn Percentage',fontsize=12,fontweight='bold')
plt.show()
```

#### Churn by Active Member

```python
active_churn = df.groupby('IsActiveMember')['Exited'].mean() * 100
print(active_churn)
```

```python
sns.barplot(x='IsActiveMember', y='Exited', data=df,palette=["red","green"])
plt.title('Churn by Active Member',fontsize=14,fontweight='bold')
plt.show()
```

#### Churn by Number of Products

```python
product_churn = df.groupby('NumOfProducts')['Exited'].mean() * 100
print(product_churn)
```

```python
sns.barplot(x='NumOfProducts', y='Exited', data=df,color="black",edgecolor="yellow")
plt.title('Churn by Number of Products')
plt.show()
```

## COHORT SEGMENTATION

#### Create Age Groups

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18,30,45,60,100],
    labels=['18-30','31-45','46-60','60+']
)
```

```python
df[['Age','AgeGroup']].head()
```

#### Churn by Age Group

```python
age_churn = df.groupby('AgeGroup')['Exited'].mean() * 100
age_churn
```

```python
age_churn.plot(kind='bar',color="yellow",edgecolor="black")
plt.title('Churn by Age Group',fontsize=14,fontweight='bold')
plt.ylabel('Churn Percentage')
plt.show()
```

#### Tenure Cohort Analysis

```python
tenure_churn = df.groupby('Tenure')['Exited'].mean() * 100
print(tenure_churn)
```

```python
tenure_churn.plot(kind='line', marker='o',markeredgecolor="red",color="black")
plt.title('Churn by Tenure',fontsize=14,fontweight='bold')
plt.ylabel('Churn Percentage')
plt.show()
```

## MODIFIED RFM ANALYSIS

#### Create Customer Value Score

```python
df['CustomerValueScore'] = (df['Balance'] * 0.4 +df['NumOfProducts'] * 0.3 + df['IsActiveMember'] * 0.3)
```

```python
df[['Balance','NumOfProducts','IsActiveMember','CustomerValueScore']].head()
```

#### Segment Customers

```python
df['CustomerSegment'] = pd.qcut(df['CustomerValueScore'],q=3,labels=['Low Value','Medium Value','High Value'])
```

```python
df[['CustomerValueScore','CustomerSegment']].head()
```

#### Churn by Customer Segment


```python
segment_churn = df.groupby('CustomerSegment')['Exited'].mean() * 100
segment_churn
```

```python
sns.barplot(x='CustomerSegment', y='Exited', data=df,palette=["red","yellow","green"])
plt.title('Churn by Customer Segment',fontsize=14,fontweight='bold')
plt.show()
```

### BUSINESS INSIGHTS

>Inactive customers showed a significantly higher churn rate compared to active members.

>Customers with only one product were more likely to leave the bank.

>High-balance customers also contributed to churn, indicating the need for premium retention strategies.

>Churn behavior varied across different geographic regions, with some regions showing higher customer loss.

>Older customers demonstrated a higher tendency to churn compared to younger age groups.

>Customers with multiple banking products had better retention and long-term engagement.

>Customer activity, product usage, and account balance were identified as major factors influencing retention.

### RECOMMENTATION

>Introduce personalized engagement campaigns to reduce churn among inactive customers.

>Encourage customers to use multiple banking products through cross-selling strategies.

>Develop loyalty programs and premium services for high-value customers.

>Implement region-specific retention strategies based on geographic churn patterns.

>Provide targeted financial solutions for older customer segments to improve satisfaction.

>Monitor high-risk customers regularly using customer behavior and activity metrics.

>Use data-driven retention strategies to improve customer lifetime value and reduce churn.
